```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
conexion=sqlite3.connect('sample.db')
cursor=conexion.cursor()
consulta='select * from customers'
cursor.execute(consulta)
resultado=cursor.fetchall()
todo_df=pd.DataFrame(resultado)
print(todo_df)
```

```python
cursor=conexion.cursor()
consulta='SELECT Firstname,PostalCode,City,Country from "customers" order by City desc'
cursor.execute(consulta)
resultado2=cursor.fetchall()
ciudades_df=pd.DataFrame(resultado2)
print(ciudades_df)
```

Con el kid="pie" le estampos indicando que es un grafico circular, con labels=ciudades_df[0] le decimos que coja la primera columna de nuestr aconsultaque en este caso son los paises

```python
cursor=conexion.cursor()
consulta='select Country, count(*) as total from customers group by Country order by total desc limit 5'
cursor.execute(consulta)
resultado2=cursor.fetchall()
ciudades_df=pd.DataFrame(resultado2)
print(ciudades_df)
ciudades_df[1].plot(kind="pie",labels=ciudades_df[0],ylabel='')
```

Usamos el scatter para indecar que va a ser un grafico de puntos

```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
conexion=sqlite3.connect('sample.db')
cursor=conexion.cursor()
consulta2='select Country, count(*) as total from customers group by Country order by total desc limit 5'
cursor.execute(consulta2)
resultado2=cursor.fetchall()
df=pd.DataFrame(resultado2)
print(df)

fig,ax = plt.subplots()
ax.scatter(x = df[0], y = df[1])
plt.xlabel("Países")
plt.ylabel("Total")

plt.show()
```
